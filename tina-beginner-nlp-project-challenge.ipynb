```python
def concat_all_text(data,column):
    text = data[column].str.cat(sep=' ')

    # removing URLs and '&amp' substrings using regex
    import re
    url_reg  = r'[a-z]*[:.]+\S+'
    text   = re.sub(url_reg, '', text)
    noise_reg = r'\&amp'
    text   = re.sub(noise_reg, '', text)
#     regex = r"[^0-9a-zA-Z\']+"
    regex = r"[^a-zA-Z\']+"
    subst = " "
    text = re.sub(regex,subst,text,0)
    
    return text

KenJee_all_comments = concat_all_text(data_kenjee_corr,'comment')
KenJee_all_comments
```
